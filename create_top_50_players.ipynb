```python
import pandas as pd
import numpy as np
from pathlib import Path

source_folder = Path("atp_transformed/2000-2024_career_averages.csv")
```

```python
df = pd.read_csv(source_folder)
```

```python
df_top_players = df[df["best_rank"] <= 50]
```

```python
# view the dataframe
print(df_top_players)
print(df_top_players.columns)
```

```python
# create the subsection based on player name:
df_player_names = df_top_players[["player_id","best_rank","worst_rank"]]

# sort
df_player_names = df_player_names.sort_values("best_rank")
```

```python
# view the created df
df_player_names
```

```python
# load the base df
df_2000_2024_with_covid = pd.read_csv("atp_transformed/2000-2024 with covid.csv")
```

```python
# merge the dfs to get the best players
big_top_players = df_player_names.merge(
    df_2000_2024_with_covid,
    how="left",
    on="player_id"
)
```

```python
# show the table contents
big_top_players
```

```python
# save the smaller dataset
big_top_players.to_csv("atp_transformed/2000-2024_rank_50_peak.csv",index=False)
```
